```python
%tensorflow_version 1.x

from sklearn.metrics.pairwise import cosine_similarity
import tensorflow_hub as hub
import tensorflow as tf
 
elmo = hub.Module("https://files.deeppavlov.ai/deeppavlov_data/elmo_ru-wiki_600k_steps.tar.gz", trainable=True)
```

```python
def simm(input_list):
  elmo_embeddings=[]
  with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    sess.run(tf.tables_initializer())
    for i in range(len(input_list)):
      embeddings=elmo([input_list[i]], signature="default", as_dict=True)["elmo"]
      v = sess.run(tf.reduce_mean(embeddings,1))[0]
      print(v.shape)
      elmo_embeddings.append(v)

  return cosine_similarity(elmo_embeddings, elmo_embeddings)
```

```python
%tensorflow_version 1.x

from sklearn.metrics.pairwise import cosine_similarity
import tensorflow_hub as hub
import tensorflow as tf
 
elmo = hub.Module("https://files.deeppavlov.ai/deeppavlov_data/elmo_ru-wiki_600k_steps.tar.gz", trainable=True)
```

```python
import numpy as np
from tqdm import tqdm
import gc

def save_elmo_embs(input_list, id_list):
  if len(input_list) != len(id_list):
    print("Incorrect lengths")
    return
  with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    sess.run(tf.tables_initializer())
    for i in tqdm(range(len(input_list))):
      embeddings = elmo([input_list[i]], signature="default", as_dict=True)["elmo"]
      v = sess.run(tf.reduce_mean(embeddings, 1))[0]
      np.save("/content/drive/MyDrive/Elmo_npys/" + str(id_list[i]) + ".npy", v)
```

```python
import pandas as pd
import docx2txt
import os

df = pd.read_csv("/content/drive/MyDrive/ids.csv")
id_list = df["ID"].to_list()

input_list = []
for path in tqdm(df["Name"]):
  input_list.append(docx2txt.process("/content/drive/MyDrive/Word/" + path))
start = len(os.listdir("/content/drive/MyDrive/Elmo_npys"))
print(start)
save_elmo_embs(input_list[start:], id_list[start:])
```
