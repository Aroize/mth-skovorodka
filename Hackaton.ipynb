```python
import textstat
import docx2txt
from tqdm import tqdm
import pandas as pd

textstat.textstat.set_lang("ru_RU")

df = pd.read_csv("/content/drive/MyDrive/ids.csv")
for i in tqdm(range(len(df))):
    path = df["Name"].iloc[i]
    score = textstat.textstat.flesch_reading_ease(docx2txt.process("/content/drive/MyDrive/Word/" + str(path)))
    print(i, score)
```

```python
from ruts import ReadabilityStats
from statistics import mean


df = pd.read_csv("/content/drive/MyDrive/ids.csv")
ari = []
for i in range(len(df)):
    path = df["Name"].iloc[i]
    rs = ReadabilityStats(docx2txt.process("/content/drive/MyDrive/Word/" + str(path)))
    stats = rs.get_stats()
    ari.append(stats["automated_readability_index"])
    print(i, stats["flesch_kincaid_grade"], stats["automated_readability_index"], 
          round(mean([stats["flesch_kincaid_grade"], stats["automated_readability_index"]]), 2))
```

```python
print(min(ari) - max(ari))
```

```python
import numpy as np
simm = np.load("/content/drive/MyDrive/simm_matrix.npy")
```

```python
json += "\"edges\": {"
for i in range(len(df)):
  row = df.iloc[i]
  n = row["Name"].split("/")[-1].split(".")[0]
  json += '"' + n + '": {'
  rec_ids = (-simm[i]).argsort()[1:3]
  recs = []
  for rec_id in rec_ids:
    p = df.iloc[rec_id]["Name"].split("/")[-1].split(".")[0]
    recs.append(p)
  for rec in recs:
    json +=  '"' + rec + '": {}, '
  json = json[:-2]
  json += "}, "
  
json = json[:-2]
json += "}}"
json
```
